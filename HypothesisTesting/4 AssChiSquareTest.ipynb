**Association between Device Type and Customer Satisfaction**\
**QUESTION:**\
**Background:**\
Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices: Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the type of device purchased and the customer's satisfaction level.\
**Data Provided:**\
The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices:\
|Satisfaction    |	Smart Thermostat |	Smart Light	| Total |\
|Very Satisfied	 |50	|70	    |120 |\
|Satisfied	     |80	|100    |180 |\
|Neutral	     |60	|90	    |150 |\
|Unsatisfied	 |30	|50	    |80  |\
|Very Unsatisfied|20	|50	    |70  |\
|Total	         |240	|360	|600 |\
**Objective:**\
To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.

**ANSWER:**\
To determine whether device type and customer satisfaction level are associated, we use the Chi-Square Test of Independence.

**Manual Chi-Square Test Calculation**

```python
import numpy as np

# Observed frequencies
observed = np.array([
    [50, 70],
    [80, 100],
    [60, 90],
    [30, 50],
    [20, 50]
])
```

```python
# Row totals, column totals, grand total
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()
```

In Python's numpy and pandas libraries, the **.sum(axis=0) function** calculates the sum of elements along the first axis, which corresponds to performing a column-wise sum.

```python
# Compute expected frequencies
expected = np.outer(row_totals, col_totals) / grand_total
```

The **numpy.outer() function** computes the outer product of two vectors. The outer product of an M-dimension vector a and an N-dimension vector b is an MxN matrix where each element out[i, j] is the product of a[i] and b[j].

numpy.outer(a, b, out = None)\
a, b: The input vectors (array_like). If the inputs are not already 1-dimensional, they are automatically flattened.\
out: An optional ndarray where the result is stored. The shape must be (M, N).

```python
print("\nExpected Frequencies:\n", expected)
```

```python
# Compute Chi-square statistic manually
chi_square = ((observed - expected) ** 2 / expected).sum()
```

```python
print("Manual Chi-Square Statistic:", chi_square)
```

**Using scipy.stats**

```python
import numpy as np
from scipy.stats import chi2_contingency
```

The **scipy.stats.chi2_contingency** function performs a chi-square test of independence on an observed contingency table in Python.\
You pass your contingency table (a 2D array or list of lists) to the chi2_contingency function.\
The function returns four values in a tuple:\
chi2_statistic: The calculated chi-square test statistic.\
p_value: The p-value of the test, used to determine statistical significance (typically compared to an alpha level of 0.05).\
degrees_of_freedom: The degrees of freedom for the test.\
expected_frequencies: An array of the expected frequencies if the variables were independent.

```python
# Observed frequency table
# Rows: Satisfaction levels
# Columns: [Smart Thermostat, Smart Light]

observed = np.array([
    [50, 70],   # Very Satisfied
    [80, 100],  # Satisfied
    [60, 90],   # Neutral
    [30, 50],   # Unsatisfied
    [20, 50]    # Very Unsatisfied
])
```

```python
# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(observed)
```

```python
print("Chi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:\n", expected)
```

```python
# Decision Rule (α = 0.05)

alpha = 0.05

if p < alpha:
    print("Reject Null Hypothesis: There is an association.")
else:
    print("Fail to Reject Null Hypothesis: No significant association between the type of smart home device purchased and the customer satisfaction level.")
```
