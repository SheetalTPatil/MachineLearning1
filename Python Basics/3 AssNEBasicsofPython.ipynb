### **Exercise 1:**  Prime Numbers
Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

The Python **range() function** generates an immutable sequence of integers and is commonly used for looping a specific number of times in for loops. It is a memory-efficient built-in function that generates numbers on demand, rather than storing the entire sequence in memory.

The range() function accepts up to three integer arguments: range(start, stop, step).\
start (Optional): The beginning integer of the sequence. If omitted, it defaults to 0. The start value is inclusive.\
stop (Required): The function stops before reaching this integer. The stop value is exclusive.\
step (Optional): The increment (or decrement) value between each number. If omitted, it defaults to 1. This value cannot be 0, or a ValueError will be raised.

### **Exercise 2: Product of Random Numbers**
Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

user_answer = int(input(f"What is the product of {num1} and {num2}? "))

if user_answer == num1 * num2:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {num1 * num2}.")
```

The **random.randint()** function in Python generates a random integer within a specified range, inclusive of both the start and end points. It is part of the built-in random module.

The syntax is random.randint(a, b), where a is the lower bound and b is the upper bound.\
Parameters:\
start: An integer representing the lowest possible value (inclusive).\
stop: An integer representing the highest possible value (inclusive).\
Returns: A random integer N such that start <= N <= stop.

### **Exercise 5: Check for Palindrome**
Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization\
**Example:**\
Input: "racecar"\
Expected Output: True


```python
import string

def is_palindrome(text):
    cleaned_text = ''.join(
        char.lower() for char in text if char.isalnum()
    )
    return cleaned_text == cleaned_text[::-1]

# Example
print(is_palindrome("racecar"))        
print(is_palindrome("A man, a plan, a canal: Panama"))  
```

The **lower() method** in Python is a built-in string method that converts all uppercase characters in a string to lowercase and returns a new string. The original string is not modified because strings are immutable in Python.

The **isalnum()** method in Python is a built-in string method used to check if all characters within a string are alphanumeric (either letters or numbers).
