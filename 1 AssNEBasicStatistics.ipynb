### Working of Python Library
In Python, a library is a group of modules that contain functions, classes and methods to perform common tasks like data manipulation, math operations, web scraping and more. Python libraries make coding faster, cleaner and more efficient by providing ready-to-use solutions for different domains such as data science, web development, machine learning and automation.\
When you import a library in Python, it gives access to pre-written code stored in separate modules. In simple terms instead of writing the logic for a task, you import the library that already has it.\
For example, on Windows, libraries are stored as .dll (Dynamic Link Libraries) and on Linux/macOS as .so files. When you run your code, Python automatically loads these modules and makes their functions available to use.

### Using Libraries in Python Programs
To use any library, it first needs to be imported into your Python program using the import statement. Once imported, you can directly call the functions or methods defined inside that library. You can import libraries in three main ways:

**Import the entire library:** import library_name. For example, import math\
**Import a specific function or class:** from library_name import function_name. For example, from math import sqrt\
**Import a library with an alias:** import library_name as alias. For example, import pandas as pd

### **3. Standardization of Numerical Variables**

```python

# To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

salesData_standardized = salesData.copy()
for col in numerical_cols:
    mu = salesData[col].mean()
    sigma = salesData[col].std()
    salesData_standardized[col] = (salesData[col] - mu) / sigma

# Comparison Plot (Before and After Standardization for one column as an example, or all)
# Let's just plot 'Total Sales Value' as an example of transformation
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
sns.histplot(salesData['Total Sales Value'], kde=True, ax=axes[0])
axes[0].set_title('Total Sales Value (Original)')
sns.histplot(salesData_standardized['Total Sales Value'], kde=True, ax=axes[1])
axes[1].set_title('Total Sales Value (Standardized)')
plt.tight_layout()
plt.savefig('standardization_comparison.png')
```

**Data Preprocessing**\
**Standardization (Z-Score Normalization)**\
Standardization transforms features to have a mean of 0 and a standard deviation of 1 using the formula: $z = \frac{x - \mu}{\sigma}$
 
**Purpose:** Many machine learning algorithms (like K-Means or SVM) are sensitive to the scale of data. Standardization ensures that a feature with a large range (like Total Sales Value) does not dominate a feature with a small range (like Volume).

**Result:** The shape of the distribution remains the same, but the x-axis is scaled.

### **4. Conversion of Categorical Data into Dummy Variables**

```python
# One-Hot Encoding
#To transform categorical variables into a format that can be provided to ML algorithms.
# Choosing meaningful categorical columns for encoding (Day, BU, Brand)
# But instructions say "categorical columns".

cols_to_encode = ['Day', 'BU', 'Brand', 'City'] 
# Actually, I'll encode all categorical columns identified earlier except Date which is a string but represents time.
#cols_to_encode = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']
salesData_encoded = pd.get_dummies(salesData, columns=cols_to_encode)

print("\nShape after encoding:", salesData_encoded.shape)
print("\nFirst few columns of encoded dataset:")
print(salesData_encoded.iloc[:, :15].head()) # Displaying first 15 columns to show new dummy columns
print("\nDummy columns example (first 10 from the end):")
print(salesData_encoded.iloc[:, -10:].head())

# Save the transformed datasets to CSV
salesData_standardized.to_csv('standardized_data.csv', index=False)
salesData_encoded.to_csv('one_hot_encoded_data.csv', index=False)
```


**One-Hot Encoding**\
Categorical data (like Brand or BU) cannot be used directly in mathematical models. One-hot encoding creates a new binary column for each unique category. That is why we convert categorical data into dummy variables so that to use for analysis purpose further.

**Impact:** The original 13 columns expanded to **87 columns** after encoding variables like SKU, Brand, and Model.

**Importance:** This allows the model to treat each category independently without assuming an artificial numerical order (which would happen with simple label encoding).
